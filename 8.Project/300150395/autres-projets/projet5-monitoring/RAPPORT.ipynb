## Objectif
Ce projet vérifie la disponibilité de plusieurs sites web et mesure leur temps de réponse.
Un script Bash appelle curl pour chaque site, et un script Python analyse les résultats.

## Résultats obtenus

| Site | Code HTTP | Temps | Statut |
|---|---|---|---|
| https://google.com | 301 | 339ms | ✅ Redirection |
| https://github.com | 200 | 489ms | ✅ OK |
| https://collegeboreal.ca | 200 | 218ms | ✅ OK |
| https://example.com | 200 | 190ms | ✅ OK |
| https://httpstat.us/404 | 000 | 236ms | ❌ Timeout |

```python
import matplotlib.pyplot as plt

sites = ['google.com', 'github.com', 'collegeboreal.ca', 'example.com', 'httpstat.us/404']
temps = [339, 489, 218, 190, 236]
statuts = ['301', '200', '200', '200', '000']
couleurs = ['orange', 'green', 'green', 'green', 'red']

plt.figure(figsize=(10, 5))
bars = plt.bar(sites, temps, color=couleurs)
plt.title('Temps de réponse des sites web (ms)')
plt.xlabel('Sites')
plt.ylabel('Temps (ms)')
plt.xticks(rotation=20)
for bar, code in zip(bars, statuts):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5, code, ha='center')
plt.tight_layout()
plt.savefig('output/monitoring.png')
plt.show()
print('Graphique sauvegardé !')
```

## Analyse

- **3/5 sites** répondent avec un code 200 (OK)
- **1/5 site** retourne un 301 (redirection normale — google.com)
- **1/5 site** est inaccessible depuis la VM (code 000 = timeout)
- Temps moyen de réponse : **294.4 ms**
- Site le plus rapide : **example.com** (190ms)
- Site le plus lent : **github.com** (489ms)

## Conclusion
Le script de monitoring fonctionne correctement. Le code 000 indique que httpstat.us/404
est inaccessible depuis le réseau de la VM, ce qui est normal en environnement cloud restreint.
