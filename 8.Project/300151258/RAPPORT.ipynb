```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Go to project folder
os.chdir(r"C:\Users\PC\Developer\INF1102-201-26H-05\8.Project\300151258")

# Load data
df = pd.read_csv("data/prix_energie.csv")

# Convert date
df["date"] = pd.to_datetime(df["date"])

# ===== ANALYSIS =====
print("\n=== ESSENCE TORONTO ===")
print("Prix moyen :", round(df["essence_toronto"].mean(), 2))
print("Prix minimum :", df["essence_toronto"].min())
print("Prix maximum :", df["essence_toronto"].max())

print("\n=== BRENT ===")
print("Prix moyen :", round(df["brent"].mean(), 2))
print("Prix minimum :", df["brent"].min())
print("Prix maximum :", df["brent"].max())

# ===== GRAPH =====
plt.figure(figsize=(10,5))

plt.plot(df["date"], df["essence_toronto"], label="Essence Toronto")
plt.plot(df["date"], df["brent"], label="Brent")

plt.title("Évolution des prix")
plt.xlabel("Date")
plt.ylabel("Prix")

plt.legend()
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Normalize data
df_norm = df.copy()
df_norm["essence_toronto"] = df["essence_toronto"] / df["essence_toronto"].max()
df_norm["brent"] = df["brent"] / df["brent"].max()

plt.figure(figsize=(10,5))
plt.plot(df["date"], df_norm["essence_toronto"], label="Essence (normalisé)")
plt.plot(df["date"], df_norm["brent"], label="Brent (normalisé)")

plt.title("Comparaison normalisée des prix")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df["difference"] = df["brent"] - df["essence_toronto"]

plt.figure(figsize=(10,5))
plt.plot(df["date"], df["difference"], color="green")
plt.title("Différence entre Brent et Essence")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Afin d’améliorer l’analyse, plusieurs graphiques ont été ajoutés :
- un graphique normalisé pour comparer les tendances
- des graphiques séparés pour plus de clarté
- un graphique de différence pour observer l’écart entre les deux prix
