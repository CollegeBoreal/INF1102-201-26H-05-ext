## Objectif
Ce projet consiste à surveiller la disponibilité de plusieurs sites web, mesurer leur temps de réponse et générer un rapport automatique.

## 1. Chargement des données

```
import matplotlib.pyplot as plt

sites = ['google.com', 'github.com', 'wikipedia.org', 'youtube.com', 'amazon.com']
temps = [419, 370, 566, 430, 294]
statuts = ['EN LIGNE', 'HORS LIGNE', 'EN LIGNE', 'EN LIGNE', 'HORS LIGNE']
codes = [200, 301, 200, 200, 503]

print('Sites analysés:', len(sites))
print('En ligne:', statuts.count('EN LIGNE'))
print('Hors ligne:', statuts.count('HORS LIGNE'))
```

## 2. Graphique des temps de réponse

```
couleurs = ['green' if s == 'EN LIGNE' else 'red' for s in statuts]

plt.figure(figsize=(10, 5))
plt.bar(sites, temps, color=couleurs)
plt.title('Temps de réponse par site (ms)')
plt.xlabel('Sites web')
plt.ylabel('Temps (ms)')
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig('../output/graphique.png')
plt.show()
print('Graphique sauvegardé !')
```

## 3. Analyse des résultats

| Site | Code HTTP | Temps (ms) | Statut |
|------|-----------|------------|--------|
| google.com | 200 | 419ms | ✅ EN LIGNE |
| github.com | 301 | 370ms | ❌ HORS LIGNE |
| wikipedia.org | 200 | 566ms | ✅ EN LIGNE |
| youtube.com | 200 | 430ms | ✅ EN LIGNE |
| amazon.com | 503 | 294ms | ❌ HORS LIGNE |

## 4. Conclusion
- **3 sites** sont en ligne sur 5
