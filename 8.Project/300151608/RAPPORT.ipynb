## 🎯 Objectif du projet

Ce projet a pour objectif d’analyser les performances de plusieurs sites web en utilisant un script Bash et Python. 
Il permet de détecter les erreurs, mesurer les temps de réponse et générer un rapport automatique.


```python
import pandas as pd
import matplotlib.pyplot as plt

data = []

with open('output/rapport.txt', 'r', encoding='utf-8') as f:
    for line in f:
        if 'http' in line and '|' in line:
            parts = [p.strip() for p in line.split('|')]
            data.append({
                'date': parts[0],
                'url': parts[1],
                'code': parts[2],
                'temps_ms': int(parts[3].replace('ms','').strip()),
                'etat': parts[4]
            })

df = pd.DataFrame(data)
df
```

```python
print("Total sites vérifiés :", len(df))
print("Sites OK (200) :", len(df[df['code'] == '200']))
print("Sites en erreur :", len(df[df['code'] != '200']))
print("Temps moyen :", round(df['temps_ms'].mean(),2), "ms")
print("Temps minimum :", df['temps_ms'].min(), "ms")
print("Temps maximum :", df['temps_ms'].max(), "ms")
```

```python
plt.figure(figsize=(8,4))
plt.plot(df['url'], df['temps_ms'], marker='o')
plt.title('Temps de réponse des sites web')
plt.xlabel('Sites')
plt.ylabel('Temps (ms)')
plt.xticks(rotation=30)
plt.grid()
plt.tight_layout()
plt.show()
```

## 📈 Analyse

Les résultats montrent que certains sites répondent plus rapidement que d'autres.
Les erreurs détectées permettent d’identifier des problèmes de connexion ou de disponibilité.
Cette analyse est utile pour améliorer la surveillance des services web.
