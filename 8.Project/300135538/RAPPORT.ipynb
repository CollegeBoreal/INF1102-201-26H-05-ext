## Projet : Monitoring de sites web avec Python et PowerShell

Ce projet permet de vérifier la disponibilité de plusieurs sites web et de mesurer leur temps de réponse.

## Méthode

Le projet fonctionne en plusieurs étapes :

- Lecture du fichier contenant les sites web
- Envoi de requêtes HTTP vers chaque site
- Récupération du code de statut
- Mesure du temps de réponse
- Génération d’un rapport et d’un graphique

```python
import pandas as pd

df = pd.read_csv("output/resultats.csv")
df
```

```python
import matplotlib.pyplot as plt

plt.bar(df["site"], df["temps_reponse"])
plt.xticks(rotation=30)
plt.title("Temps de réponse des sites web")
plt.show()
```

```python
df["disponibilite"].value_counts()
```

```python
## Analyse

Les résultats montrent que :

- Les sites avec un temps de réponse faible sont plus rapides
- Les sites avec statut 200 sont disponibles
- Les sites indisponibles peuvent avoir des problèmes réseau ou serveur

Ce type d’analyse permet de surveiller la performance des sites web.
```

## Conclusion

Ce projet montre comment combiner PowerShell et Python pour automatiser une tâche réelle.
