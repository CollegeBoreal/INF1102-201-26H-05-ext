```python
## Objectifs du projet

Les objectifs principaux de ce projet sont :

- Automatiser la récupération des données d’une cryptomonnaie via une API
- Stocker les données de manière persistante
- Générer un graphique mis à jour à chaque exécution
- Utiliser PowerShell et Python de manière complémentaire
- Exploiter Jupyter Notebook pour l’analyse et la présentation des résultats
```

```python
## Structure du projet

La structure du projet est la suivante :

- `scripts/analyse.ps1` : script PowerShell principal
- `scripts/analyse.py` : script Python de traitement
- `data/prices.csv` : fichier d’historique des prix
- `output/graph.png` : graphique généré automatiquement
- `output/rapport.txt` : rapport texte généré
- `RAPPORT.ipynb` : analyse et présentation des résultats

Cette organisation permet une séparation claire entre les scripts,
les données d’entrée et les résultats.
```

```python
## Fonctionnement général

L’exécution du projet suit les étapes suivantes :

1. Le script PowerShell `analyse.ps1` est lancé
2. Il appelle le script Python `analyse.py`
3. Le script Python récupère le prix actuel du Bitcoin via une API
4. Le prix est ajouté à un fichier CSV contenant l’historique
5. Un graphique est généré à partir de l’ensemble des données
6. Les résultats sont enregistrés automatiquement

Le notebook Jupyter n’exécute pas directement la logique principale,
mais sert d’interface d’analyse et de démonstration.
```
