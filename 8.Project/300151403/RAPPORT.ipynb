# 🌦️ RAPPORT MÉTÉO AUTOMATISÉ

Analyse des données météo récupérées via API OpenWeatherMap.

```python
import json

data = json.load(open("data/meteo.json"))

data
```

```python
temp = data["main"]["temp"]
humidity = data["main"]["humidity"]
pressure = data["main"]["pressure"]

print("Température :", temp)
print("Humidité :", humidity)
print("Pression :", pressure)
```

```python
import matplotlib.pyplot as plt

plt.bar(
    ["Temp", "Humidity", "Pressure"],
    [temp, humidity, pressure]
)
plt.title("Analyse météo")
plt.show()
```
