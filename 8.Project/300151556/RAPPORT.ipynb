## 🎯 Objectif
Ce projet consiste à analyser un fichier log à l'aide de Bash et Python afin de générer un rapport automatique contenant les statistiques principales.

## ⚙️ Méthode

1. Le script Bash lance l'analyse
2. Le script Python lit le fichier `sample.log`
3. Extraction des données :
   - Adresses IP
   - URLs
4. Calcul des fréquences
5. Génération du fichier `rapport.txt`

```python
from collections import Counter

file = "data/sample.log"

ips = []
urls = []

with open(file, 'r') as f:
    for line in f:
        parts = line.split()
        ips.append(parts[0])
        urls.append(parts[5])

top_ips = Counter(ips).most_common(3)
top_urls = Counter(urls).most_common(3)

print("Top IPs :", top_ips)
print("Top URLs :", top_urls)
```

## 📈 Résultats

Les adresses IP les plus fréquentes sont identifiées ainsi que les pages les plus consultées.

Cela permet de comprendre :
- le trafic utilisateur
- les pages les plus utilisées

```python
import matplotlib.pyplot as plt

labels = [ip for ip, _ in top_ips]
values = [count for _, count in top_ips]

plt.bar(labels, values)
plt.title("Top IPs")
plt.xlabel("IP")
plt.ylabel("Nombre de requêtes")
plt.show()
```
