```python
# Somme des termes d'une suite arithmétique
# Formule : S = n/2 * (premier terme + dernier terme)

# Saisie des valeurs
a = float(input("Entre le premier terme (a) : "))
d = float(input("Entre la raison (d) : "))
n = int(input("Entre le nombre de termes (n) : "))

# Calcul du dernier terme
dernier_terme = a + (n - 1) * d

# Calcul de la somme
S = (n / 2) * (a + dernier_terme)

# Affichage de tous les termes
print("\n--- Termes de la suite ---")
for i in range(n):
    terme = a + i * d
    print(f"  U{i+1} = {terme}")

# Résultat final
print(f"\n--- Résultat ---")
print(f"Dernier terme     : {dernier_terme}")
print(f"Somme des {n} termes : {S}")
```
